### **1.3. Search Algorithm Terminologies**

•  **Search**: The process of finding something in a particular search space by following a set of steps. Generally, there are three main factors to consider when searching:
1.	**Search Space:** A search space represents a range of possible solutions that may be available to a given system.
2.	**Start State:** This is where an agent begins the search.
3.	**Goal test:** It's function that determine whether the goal state has been achieved based on the current state.

•	**Search tree:** In search, it refers to a tree representation of the search problem. In a search tree, the root node corresponds to the initial state.

**•	Actions:** It describes all the agents' available actions.

•	**Path Cost:** A path's cost is assigned numerically by this function.

•	**Solution:** A sequence of actions leading from a start node to a goal node.

•	**Optimal Solution:** The ideal solution with the lowest cost.



---


In order to compare the efficiency of search algorithms, it is important to 
consider the following four properties:



•	**Completeness:** It is said that a search algorithm is complete when it ensures that for any random input there will be at least one solution.

•	**Optimality:** When an algorithm provide a solution with lowest path costs (highest efficiency) compared to all other solutions, it is described as an optimal solution.

•	**Time Complexity:** How long an algorithm takes to execute a single task is called the time complexity of the algorithm.

•	**Space Complexity:** This is the amount of storage space required at any given time during the search, in accordance with the complexity of the problem.

###**1.4. Types of search algorithms**
There are uninformed (Blind search) and informed (Heuristic search) search algorithms depending on the search problems.

####**-Uninformed/Blind Search:**

There is no domain knowledge in an uninformed search. For example, whether a goal is close or where it is located. Since the only information it contains is how to traverse the tree and how to identify nodes and goals. As it checks each node of the tree until it reaches the goal node without knowing anything about the search space, it is called a blind search.

####Basically, there are six different types of uninformed search:

* Breadth-first search
* Depth-first search
* Depth-limited search 
* Iterative deepening search
* Bidirectional search
* Uniform cost search

####**- Informed/Heuristic Search:**
Domain knowledge is the basis for informed search algorithms. So, when there is knowledge on the problem, it is possible to guide the search easily. Furthermore, using informed search strategies can produce more effictive results than uninformed search strategies. Furthermore, through informed search, many complex problems can be solved that would be impossible with any other strategy. Alternatively, informed searches are referred to as heuristic searches.

####There are two different types of informed search:
* Greedy search
* A* search

###**3.1. Constraint satisfaction problems (CSP)**

The goal is to find one solution that satisfies all the constraints of the search problem,Domains and variables determine the set of all possible solutions, complete or partial. 

We have set of variables, constraints and known domains of each variable. We have to assign a value to each variable so that all conditions are met.

* Set of variables {X1, X2,…,Xn}

* Set of Domains {D1, D2,…,Dn}

* Set of constraints {C1, C2,…,Cn } 

* Set of relationships {R1, R2,…,Rn}

The CSP search graph’s nodes represent the assignments, and an edge from node u to vertex v corresponds to a change in a single variable.


We use two different types of variables to form the constraint network: 

The board is represented by the first class of variables (state variables), which will also support the constraints specifying the game's rules.

The second type of variables, known as quantified variables, are a representation of the choices made by the players. They are connected to the first type of variables by channeling constraints, which ensure that the values chosen by all variables are coherent.
