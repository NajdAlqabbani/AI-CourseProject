##**2. Game as search**
Game trees represent sequential games graphically. It provides information about the players, payoffs (representing the motivations of players, such as profit, quantity, and "utility"), strategies, and the order of moves. There are nodes (vertices) in the game tree, which represent places where players can take actions, connecting these points with edges that represent the actions that can be performed at those points. As the first decision, an initial (or root) node represents that decision. A terminal node represents an end to the game as every edge from the first node passes through it. There is a payoff label at each terminal node that indicates how much each player will earn if the game ends there. Therefore, the **X-O game** can be viewed as an example:

##**3. Evaluation function**

Searching the entire search space is sometimes not feasible due to its size. In this case we need to use heuristic search. Rather than trying all possible paths to reach the goal state, heuristic search focuses on paths that seem to be moving us closer to the goal. Even though we may not always know when we are close to the goal state, a good guess might be helpful. Heuristics are used to help us make that guess. The search tree must be evaluated based on some criteria (for example, how near the node is to the target). Guesses can be made quickly with the help of this function. Moreover, the evaluation function only needs three values: Win, Lose, and Tie. The value of each inner node in the game tree is the maximum or minimum value, where the first player's win is denoted by value +1, the tie by his 0, and the second player's win by value -1, the first and second players wins value   depends on what game they are  playing ,the value of each inner node in the game tree is the maximum or minimum of its children's values ,depending on whether the first or second player is to move respectively. ##**5. Min-Max** 
The Min-Max algorithm can be used to solve decision-making problems and solve games. The algorithm provides the most optimal move for the player, taking into account how the competitor is playing as well. There are many games in Artificial Intelligence that use this algorithm, such as chess, X-O, and many other games where two players play against each other.

As part of this algorithm, two players play the game; one is known as **MAX** who tries to select the maximized value, and the other is **MIN** who tries to select the minimized value. In other words, the main aim of playing games using this algorithm is to minimize the benefits of competitors and maximize the benefits of themselves!

##**7. X-O GAME**

In this game, two players pick a mark ( X or O ) and alternately put that mark in a square of a grid consisting of nine squares (usually), but the grid may also be 4x4, or 5x5, etc. And the purpose of playing X-O is to win by getting 3 of your marks (either X or O) in a straight line. There are three ways in which a straight line can run: diagonally, vertically, or horizontally. 

The game's formulation will be as follow:

* **(Intial state)** The game is set up with an empty board.
* **(Players)** Before the game starts, each player must choose one mark (X or O) and use that mark throughout the game.

* **(Actions)** Only one mark can be made per turn by each player. Moreover,it is only possible to place a mark on empty squares. Once placed, it will stay there forever.

* **(Terminal test)** The game is over when one player win by having a straight line of three marks or the game is considered a tie when all 9 squares are marked even if no player has a straight line of three marks.


##**8. Implementation of Min-Max Algorithm** 

```python
import math
import time
 

class Game:
    def __init__(self):
        self.init_game()
 
    def init_game(self):
        self.current_state = [[' ',' ',' '],
                              [' ',' ',' '],
                              [' ',' ',' ']]
        self.depth=0
        self.size=0
                             
        # X is start first
        self.player_turn = 'X'
   
   
             
 
    def make_board(self):
        for i in range(0, 3):
            for j in range(0, 3):
                print('{}|'.format(self.current_state[i][j]), end=" ") #inter the position
            print()
        print()
 
        #_______________________________________________________________
 
       
    # Check the move is legal
    def check_valid(self, px, py):#px = position x , py = position y
        if px < 0 or px > 2 or py < 0 or py > 2:
            return False
        elif self.current_state[px][py] != ' ':
            return False
        else:
            return True
        #________________________________________________________________
    def check_winner(self):
    # Check Vertical
        for i in range(0, 3):#012 (3 not includ)
            if (self.current_state[0][i] != ' ' and
                self.current_state[0][i] == self.current_state[1][i] and
                self.current_state[1][i] == self.current_state[2][i]):
 
                return self.current_state[0][i]  # e.g. p1=p4 & p4=p7 - return the sign in this position
 
    # Check Horizontal
        for i in range(0, 3):
            if (self.current_state[i] == ['X', 'X', 'X']):
                return 'X'
            elif (self.current_state[i] == ['O', 'O', 'O']):
               
                return 'O'  
 
    # Check first diagonal
        if (self.current_state[0][0] != ' ' and
            self.current_state[0][0] == self.current_state[1][1] and
            self.current_state[0][0] == self.current_state[2][2]):
           
            return self.current_state[0][0] # e.g.p1=p5 & p5=p9 - return the sign in this position
 
    # Check second diagonal
        if (self.current_state[0][2] != ' ' and
            self.current_state[0][2] == self.current_state[1][1] and
            self.current_state[0][2] == self.current_state[2][0]):
           
            return self.current_state[0][2]  # e.g.p3=p5 & p5=p7 - return the sign in this position
 
    # Full or not
        for i in range(0, 3):
            for j in range(0, 3):
            # There's an empty position, the game will continue
               if (self.current_state[i][j] == ' '):
                   return None 
 
    # If a tie!
       
        return ' '
 
    #__________________________________________________________
    # Player 'X' is max
    def max(self):
 
    # We're initially setting it to -infinty as the worst case:
       maxv = -math.inf  #max vlaue
       px = None
       py = None
       score = self.check_winner()
 
    # the evaluation function:
    # -1 - loss
    # 0  - a tie
    # 1  - win
       if score == 'X':  
           return (1, 0, 0)
       elif score == 'O':
           return (-1, 0, 0)
       elif score == ' ':
           return (0, 0, 0)
 
       self.size +=1 #size of tree

       for i in range(0, 3):
           for j in range(0, 3):
 
                # On the empty position player 'X' makes a move and calls Min
                # That's one branch of the game tree.
               if self.current_state[i][j] == ' ':
                  self.current_state[i][j] = 'X'
                 
                  (m, min_i, min_j) = self.min() #Call the min
                 
                # Change the maxv value if needed
                  if m > maxv:
                      maxv = m
                      px = i
                      py = j
                   
                # Setting back the position to empty
                  self.current_state[i][j] = ' '
                           
       return (maxv, px, py)
 
#___________________________
# Player 'O' is min, in this case AI
    def min(self):
 
    # We're initially setting it to Infinty as the worst case:
       minv = math.inf # min value
       qx = None
       qy = None
       score = self.check_winner()
 
    # the evaluation function:
    # -1 - win
    # 0  - a tie
    # 1  - loss
       if score == 'X':
           return (1, 0, 0)
       elif score == 'O':
           return (-1, 0, 0)
       elif score == ' ':
           return (0, 0, 0)
 
       self.size +=1 #size of tree

       for i in range(0, 3):
           for j in range(0, 3):

               # On the empty position player 'O' makes a move and calls Max
                # That's one branch of the game tree.
               if self.current_state[i][j] == ' ':
                   self.current_state[i][j] = 'O'

                   (m, max_i, max_j) = self.max() #Calls the max

                   # Change the minv value if needed
                   if m < minv:
                       minv = m
                       qx = i
                       qy = j

                   # Setting back the position to empty  
                   self.current_state[i][j] = ' '
                   
       return (minv, qx, qy)
 
    #______________________________________________
    #Start the game
    def play(self):

        # always true until the end of the game
        while True:
            self.make_board()
            self.score = self.check_winner()
 
        # Printing the right message if the game has ended
 
            if self.score != None: #the game is end
                
                if self.score == 'X':
                    print('The winner is X!')
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
                elif self.score == 'O':
                    print('The winner is O!')
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
                elif self.score == ' ':
                    print("It's a tie!")
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
 
                self.init_game()
                return
 
        # If X player turn
            if self.player_turn == 'X':
                self.depth +=1

                while True:
                   
                    start = time.time()
                    (m,qx,qy) = self.max()
                    end = time.time()

                    print('Evaluation time: {}s'.format(round(end - start, 5))) #num - digit
                    print('Recommended move: X = {}, Y = {}'.format(qx, qy))
 
                    px = int(input('Enter the X position: '))
                    py = int(input('Enter the Y position: '))
 
                    (qx, qy) = (px, py)
 
                    if self.check_valid(px, py):
                        self.current_state[px][py] = 'X'
                        self.player_turn = 'O'
                        break
                    else:
                        print('The move is not valid! Try again.')
 

        # If it's O player turn
            else:
 
                self.depth +=1
 
                start = time.time()
                (m, px, py) = self.min()
                end = time.time()
 
                print('Evaluation time for AI: {}s'.format(round(end - start, 5))) #num - digit0
 
                self.current_state[px][py] = 'O'
                self.player_turn = 'X'
            #_________________________________________________
 

#main        
if __name__ == "__main__":
           
            print("select position (XY):")
            print("00| 01| 02|")
            print("10| 11| 12|")
            print("20| 21| 22|")
            print()
            g = Game()
            g.play()  
 
```

```python
import math
import time
 

class Game:
    def __init__(self):
        self.init_game()
 
    def init_game(self):
        self.current_state = [[' ',' ',' '],
                              [' ',' ',' '],
                              [' ',' ',' ']]
        self.depth=0
        self.size=0
                             
        # X is start first
        self.player_turn = 'X'
   
   
             
 
    def make_board(self):
        for i in range(0, 3):
            for j in range(0, 3):
                print('{}|'.format(self.current_state[i][j]), end=" ") #inter the position
            print()
        print()
 
        #_______________________________________________________________
 
       
    # Check the move is legal
    def check_valid(self, px, py):#px = position x , py = position y
        if px < 0 or px > 2 or py < 0 or py > 2:
            return False
        elif self.current_state[px][py] != ' ':
            return False
        else:
            return True
        #________________________________________________________________
    def check_winner(self):
    # Check Vertical
        for i in range(0, 3):#012 (3 not includ)
            if (self.current_state[0][i] != ' ' and
                self.current_state[0][i] == self.current_state[1][i] and
                self.current_state[1][i] == self.current_state[2][i]):
 
                return self.current_state[0][i]  # e.g. p1=p4 & p4=p7 - return the sign in this position
 
    # Check Horizontal
        for i in range(0, 3):
            if (self.current_state[i] == ['X', 'X', 'X']):
                return 'X'
            elif (self.current_state[i] == ['O', 'O', 'O']):
               
                return 'O'  
 
    # Check first diagonal
        if (self.current_state[0][0] != ' ' and
            self.current_state[0][0] == self.current_state[1][1] and
            self.current_state[0][0] == self.current_state[2][2]):
           
            return self.current_state[0][0] # e.g.p1=p5 & p5=p9 - return the sign in this position
 
    # Check second diagonal
        if (self.current_state[0][2] != ' ' and
            self.current_state[0][2] == self.current_state[1][1] and
            self.current_state[0][2] == self.current_state[2][0]):
           
            return self.current_state[0][2]  # e.g.p3=p5 & p5=p7 - return the sign in this position
 
    # Full or not
        for i in range(0, 3):
            for j in range(0, 3):
            # There's an empty position, the game will continue
               if (self.current_state[i][j] == ' '):
                   return None
 
    # If a tie!
       
        return ' '
 
    #__________________________________________________________
    # Player 'X' is max
    def max_alpha_beta(self, alpha, beta):
 
    # We're initially setting it to -infinty as the worst case:
       maxv = -math.inf  #max vlaue
       px = None
       py = None
       score = self.check_winner()
 
    # the evaluation function:
    # -1 - loss
    # 0  - a tie
    # 1  - win
       if score == 'X':  
           return (1, 0, 0)
       elif score == 'O':
           return (-1, 0, 0)
       elif score == ' ':
           return (0, 0, 0)
 
       self.size +=1 #size of tree

       for i in range(0, 3):
           for j in range(0, 3):
 
                # On the empty position player 'X' makes a move and calls Min
                # That's one branch of the game tree.
               if self.current_state[i][j] == ' ':
                  self.current_state[i][j] = 'X'
                 
                  (m, min_i, min_j) = self.min_alpha_beta( alpha, beta) #Call the min
                 
                # Change the maxv value if needed
                  if m > maxv:
                      maxv = m
                      px = i
                      py = j
                   
                # Setting back the position to empty
                  self.current_state[i][j] = ' '
 
                  if maxv >= beta: #alpha >= beta
                        return (maxv, px, py) #pruning
 
                  if maxv > alpha:
                        alpha = maxv
                 
                           
       return (maxv, px, py)
 
#___________________________
# Player 'O' is min, in this case AI
    def min_alpha_beta(self, alpha, beta):
 
    # We're initially setting it to Infinty as the worst case:
       minv = math.inf # min value
       qx = None
       qy = None
       score = self.check_winner()
 
    # the evaluation function:
    # -1 - win
    # 0  - a tie
    # 1  - loss
       if score == 'X':
           return (1, 0, 0)
       elif score == 'O':
           return (-1, 0, 0)
       elif score == ' ':
           return (0, 0, 0)
 
       self.size +=1 #size of tree

       for i in range(0, 3):
           for j in range(0, 3):
               # On the empty position player 'O' makes a move and calls Max
                # That's one branch of the game tree.
               if self.current_state[i][j] == ' ':
                   self.current_state[i][j] = 'O'
                   (m, max_i, max_j) = self.max_alpha_beta( alpha, beta) #Calls the max
                   # Change the minv value if needed
                   if m < minv:
                       minv = m
                       qx = i
                       qy = j

                   # Setting back the position to empty  
                   self.current_state[i][j] = ' '
 
                   if minv <= alpha: #beta <=alpha
                        return (minv, qx, qy)
 
                   if minv < beta:
                        beta = minv
                   
       return (minv, qx, qy)
 
    #______________________________________________
    #Start the game
    def play_alpha_beta(self):
      
        # always true until the end of the game
        while True:
            self.make_board()
            self.score = self.check_winner()
 
        # Printing the right message if the game has ended
 
            if self.score != None: #the game is end
                if self.score == 'X':
                    print('The winner is X!')
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
                elif self.score == 'O':
                    print('The winner is O!')
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
                elif self.score == ' ':
                    print("It's a tie!")
                    print('Depth:',self.depth)
                    print('Size:',self.size)
                   
 
                self.init_game()
                return
 
        # If X player turn
            if self.player_turn == 'X':
                self.depth +=1
                while True:
                   
                    start = time.time()
                    (m, qx, qy) =self.max_alpha_beta(-math.inf,math.inf)
                    end = time.time()
                    print('Evaluation time: {}s'.format(round(end - start, 5))) #num - digit
                    print('Recommended move: X = {}, Y = {}'.format(qx, qy))
 
                    px = int(input('Enter the X position: '))
                    py = int(input('Enter the Y position: '))
 
                    (qx, qy) = (px, py)
 
                    if self.check_valid(px, py):
                        self.current_state[px][py] = 'X'
                        self.player_turn = 'O'
                        break
                    else:
                        print('The move is not valid! Try again.')
 

        # If it's O player turn
            else:
 
                self.depth +=1
 
                start = time.time()
                (m, px, py) = self.min_alpha_beta(-math.inf,math.inf)#max
                end = time.time()
 
                print('Evaluation time for AI: {}s'.format(round(end - start, 5))) #num - digit0
 
                self.current_state[px][py] = 'O'
                self.player_turn = 'X'
            #_________________________________________________
 

#main        
if __name__ == "__main__":
           
            print("select position (XY):")
            print("00| 01| 02|")
            print("10| 11| 12|")
            print("20| 21| 22|")
            print()
            g = Game()
            g.play_alpha_beta()
```
